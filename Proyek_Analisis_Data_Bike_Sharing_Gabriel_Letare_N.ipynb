## Visualization & Explanatory Analysis

### Dalam rentang 2 tahun (2011-2012) pada jam berapakah total peminjaman sepeda tertinggi dan terendah?

```python
byhour_df = hour_df.groupby(by="hr").agg({
    "instant" : "nunique",
    "cnt": ["max", "min", "mean", "sum"]
})

byhour_df = byhour_df.reset_index()
byhour_df.columns = ['hour', 'unique_instants', 'max_cnt', 'min_cnt', 'mean_cnt', 'sum_cnt']

plt.figure(figsize=(10, 5))
plt.plot(byhour_df["hour"], byhour_df["sum_cnt"], marker='o', linewidth=2, color="#72BCD4")
plt.xticks(range(0, 24))
plt.xlabel("Jam")
plt.ylabel("Total Rental")
plt.title("Banyak Rental per Jam", loc="center", fontsize=20)
plt.show()
```

### Total peminjaman sepeda secara *casual* terbanyak terjadi pada bulan apa?

```python
casual_per_month_df = hour_df.groupby(by="mnth").agg({
    "casual": "sum",
})

casual_per_month_df = casual_per_month_df.reset_index()
colors = ["#72BCD4" if x < max(casual_per_month_df['casual']) else 'darkblue' for x in casual_per_month_df['casual']]

plt.figure(figsize=(10, 5))
bars = sns.barplot(
    x="mnth",
    y="casual",
    data=casual_per_month_df,
    color="#72BCD4"
)

bars.patches[casual_per_month_df['casual'].idxmax()].set_color('darkblue')

plt.title("Banyak Rental Secara Casual per Bulan", loc="center", fontsize=15)
plt.xlabel("Month")
plt.ylabel("Total Casual Rentals")
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

Dari visualisasi di atas terlihat bahwa peminjaman sepeda dalam rentang 2 tahun tebanyak terjadi pada bulan Juli. Tidak terdapat perbedaan yang signifikan pula antara peminjaman dari bulan Mei hingga September.
